```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
from sklearn import preprocessing
```

```python
data=pd.read_csv('/content/canada_per_capita_income.csv')
canada_df=pd.DataFrame(data)
canada_df.head()
```

```python
reg=linear_model.LinearRegression()
reg.fit(canada_df[['year']],canada_df['per capita income (US$)'])
reg.predict([[2020]])
```

```python
data=pd.read_csv('/content/hiring.csv')
hiring_df=pd.DataFrame(data)
hiring_df.head()
```

```python
hiring_df.dropna(inplace=True)
```

```python
hiring_df['experience'].unique()
```

```python
hiring_df.replace('five',5,inplace=True)
hiring_df.replace('two',2,inplace=True)
hiring_df.replace('seven',7,inplace=True)
hiring_df.replace('three',3,inplace=True)
hiring_df.replace('eleven',11,inplace=True)
```

```python
hiring_df['experience'].unique()
```

```python
model=linear_model.LinearRegression()
model.fit(hiring_df[['test_score(out of 10)','interview_score(out of 10)','experience']],hiring_df['salary($)'])
model.predict([[2,9,6]])
```

```python
model.predict([[12,10,10]])
```
